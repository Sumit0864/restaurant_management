### Determine the percentage of restaurants that offer online delivery.

```python
total_restaurants=len(df)
online_delivery_restaurants=df['Has Online delivery']=='Yes'
offline_delivery_restaurants=df['Has Online delivery']!='Yes'
percentage_online_delivery=online_delivery_restaurants.sum()/total_restaurants*100
print("The total percentage of resturants who serve online delivery are : {}%".format(percentage_online_delivery.round(2)))
```

```python
plt.figure(figsize=(10,8))
labels = ['Online Delivery', ' No Online Delivery']
sizes = [percentage_online_delivery, 100 - percentage_online_delivery]
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, colors=['lightblue', 'gold'], autopct='%2f%%', startangle=90)
plt.title('Percentage Distribution of Restaurants by Delivery Type')
plt.axis('equal')  
plt.legend()
plt.show()
```

### Compare the average ratings of restaurants with and without online delivery.

```python

avg_rating_online = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
avg_rating_offline = df[df['Has Online delivery'] != 'Yes']['Aggregate rating'].mean()

print(f"Average rating for restaurants with online delivery: {avg_rating_online:.2f}")
print(f"Average rating for restaurants without online delivery: {avg_rating_offline:.2f}")
```
